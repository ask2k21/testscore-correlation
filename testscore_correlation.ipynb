```python
def gradient_decent(x,y):
    m=b=0
    learning_rate = 0.1
    iterations = 1000
    n =len(x)
    cost_p=0
    for i in range(iterations):
        y_pred = m*x+b
        cost = (1/n)*sum(value**2 for value in (y-y_pred))
        if math.isclose(cost_p,cost,rel_tol=1e-20):
            break
        md = -(2/n)*sum(x*(y-y_pred))
        bd = -(2/n)*sum(y-y_pred)
        m = m - learning_rate * md
        b = b - learning_rate * bd
        cost_p = cost
        print("m {} b {} cost {} iteration {}".format(m,b,cost,i))
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
gradient_decent(x,y)
```
